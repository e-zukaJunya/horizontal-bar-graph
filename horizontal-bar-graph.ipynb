```python
import pandas as pd
import numpy as np
import codecs
import os
import shutil
import datetime
import math
from scipy.stats import pearsonr
from sklearn.datasets import dump_svmlight_file
from sklearn.metrics import mean_squared_error
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from numpy.random import *
from scipy import stats
from sklearn.metrics import silhouette_samples
from matplotlib import cm
from sklearn import preprocessing
```

```python
df = pd.DataFrame([[1,2,1],[3,5,1]])
df_graph = df.apply(lambda x:x/x.sum(), axis=1)
df_graph = df_graph.round(2)
df_graph
```

```python

my_plot = df_graph.plot(kind='barh', stacked=True, title="Occupancy")
my_plot.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0, fontsize=18)

column_len = len(df_graph.columns)
row_len = len(df_graph)

for i in range(row_len):
    for j in range(column_len):
        p = my_plot.patches[i*column_len + j]
        my_plot.annotate(
            p.get_width().round(2)*100,
            xy=(p.get_x() + p.get_width() / 2, p.get_y() + p.get_height() / 2),
            textcoords="offset points",
            ha='center',
            va='bottom'
            )
```
